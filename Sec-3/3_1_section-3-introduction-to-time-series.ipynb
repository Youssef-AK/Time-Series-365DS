### The QQ Plot

```python
import scipy.stats
import pylab
```

```python
scipy.stats.probplot(df_comp.spx, plot = pylab)
pylab.show()
```

## Plot-above: 
- Y-axis : The price
- X-axis : How many standard deviations away from the mean these values are.
- Red-line : What the data points should follow, if they are Normally Distributed.


#### Our data is not Normal Distributed. !
 **However** this we usually expect from time series data
